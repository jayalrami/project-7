```python
import math
import time
import random
import uuid
from datetime import datetime

print("==========================")
print("Welcome to Multi-Utility Toolkit")
print("==========================")

def datetime_menu():
    while True:
        print("\n Date and Time Operations")
        print("1. Display Current Date and Time")
        print("2. Calculate The Diffrence Between Two Dates")
        print("3. Formet Date Info Custome Formet")
        print("4. Stopwatch")
        print("5. Countdown Timer")
        print("6.Back To Main Menu")

        choice = input("\nEnter Your Choice: ")

        if choice == "1":
            now = datetime.now()
            print("\n Current Date and Time: ",now.strftime("%Y-%m-%d %H:%M:%S"))
            print("==========================")
            
        elif choice == "2":
            d1 = input("Enter the First Date(YYYY-MM-DD): ")
            d2 = input("Enter the Second Date(YYYY-MM-DD): ")
            date1 = datetime.strptime(d1, "%Y-%m-%d")
            date2 = datetime.strptime(d2, "%Y-%m-%d")
            diff = abs(date2 - date1)
            print("Diffrence : ",diff.days, "days")
            print("==========================")
            
        elif choice == "3":
            formet = input("Enter Your Customize Formet(%d-%m-%Y): ")
            print("Formated Date: ",datetime.now().strftime(formet))
            print("==========================")
            
        elif choice == "4":
            input("Press Enter To Start StoppWatch : ")
            start = time.time()
            input("Press Enter to Stop Watch : ")
            end = time.time()
            elapsed_time = round(end - start,5)
            print("Elapsed Time: ",elapsed_time, "Secoends")
            print("==========================")

        elif choice == "5":
            second = int(input("Enter The Countdown Start : "))
            while second > 0:
                print("\n Time Left : ",second,end = "s")
                time.sleep(1)
                second -= 1
            print("\n Times Up.....!")
            print("==========================")

        elif choice == "6" :
            break

def math_menu():
    while True:
        print("\nMathematical Operations")
        print("1. Calculate the Factorial")
        print("2. Compound Interest")
        print("3. Trigonometric Calculations")
        print("4. Area Of Geometric Shapes")
        print("5. Back To Main Menu")

        choice = input("\nEnter Your Choice: ")

        if choice == "1":
            n = int(input("Enter a Number:"))
            print("Factorial: ",math.factorial(n))
            print("==========================")

        elif choice == "2":
            p = float(input("Enter principal amount : "))
            r = float(input("Enter Rate Of Intrest(%) :"))
            t = float(input("Enter time to return : "))
            ci = p * (1 + r / 100) ** t
            print("Compound Intrest: ",round(ci,2))
            print("==========================")

        elif choice == "3":
            angle = float(input("Enter angle in degree : "))
            rad = math.radians(angle)
            print("Sin : ",math.sin(rad))
            print("Cos : ",math.cos(rad))
            print("Tan : ",math.tan(rad))
            print("==========================")

        elif choice == "4":
            r = float(input("Enter the Radius of Circle : "))
            print("Area Of Circle : ",math.pi * r * r)
            print("==========================")

        elif choice == "5":
            break

def random_menu():
    while True:
        print("\n Random Data Genration")
        print("1. Genrate Random Number")
        print("2. Gemrate Random List")
        print("3. Create Random Password")
        print("4. Genrate Random OTP")
        print("5. Back To Main Menu")

        choice = input("\nEnter Your Choice: ")

        if choice == "1":
            print("Random Number: ",random.randint(1,100))
            print("==========================")

        elif choice == "2":
            print("Random List: ",random.sample(range(1,50), 5))
            print("==========================")

        elif choice == "3":
            length = int(input("Enter the Password length: "))
            char = "abcdefghijklmnopqrstuvwxyzABCDEFJHIJKLMNOPQRSTUVWXYZ0123456789@#$"
            password = "".join(random.choice(char) for i in range(length))
            print("Genrate The Password: ",password)
            print("==========================")

        elif choice == "4":
            print("Genrate Random OTP: ",random.randint(100000,999999))
            print("==========================")

        elif choice == "5":
            break

def file_menu():
    while True:
        print("\n File Operations")
        print("1. Create File")
        print("2. Write File")
        print("3. Read File")
        print("4. Append The File")
        print("5. Back To Main Menu")

        choice = input("\nEnter Your Choice : ")

        if choice == "1":
            filename = input("Enter File Name: ")
            open(filename,"w").close()
            print("File Created Successfully....!")
            print("==========================")

        elif choice == "2":
            filename = input("Enter File Name: ")
            data = input("Enter The Data To Write: ")
            with open(filename,"w")as f:
                f.write(data)
            print("Data Written Successfully.....!")
            print("==========================")

        elif choice == "3":
            filename = input("Enter File Name: ")
            with open(filename,"r")as f:
                print("\n File Content: ")
                print(f.read())
                print("==========================")

        elif choice == "4":
            filename = input("Enter File Name: ")
            data = input("Enter Data To Append: ")
            with open(filename,"a")as f:
                f.write("\n" + data)
            print("Data Added Successfully......!")
            print("==========================")

        elif choice == "5":
            break

def Main_menu():
    while True:
        print("\n Welcome To Multi-Utility Toolkit")
        print("1. Datetime and Time Operations")
        print("2. Mathamatices Operations")
        print("3. Random Data Genrations")
        print("4. Genrate Unique Identifier(UUID)")
        print("5. File Operations")
        print("6. Explore Module Attributes (dir())")
        print("7. Exit")
        print("==================================")

        choice = input("\nEnter Your Choice: ")

        if choice == "1":
            datetime_menu()

        elif choice == "2":
            math_menu()

        elif choice == "3":
            random_menu()

        elif choice == "4":
            print("Genrated UUID: ",uuid.uuid4())
            print("==================================")

        elif choice == "5":
            file_menu()

        elif choice == "6":
            module_name = input("Enter Module Name: ")
            print("Avaliable Attributes: ")
            print(dir(__import__(module_name)))
            print("==================================")

        elif choice == "7":
            print("==================================")
            print("Thank You For Using The Multi-Utility Toolkit! ")
            print("==================================")
            break

if __name__ == "__main__":
    Main_menu()
    









            


```
